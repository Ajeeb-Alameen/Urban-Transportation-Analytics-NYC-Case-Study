```python
import requests
from bs4 import BeautifulSoup
import os

# URL of the TLC Trip Record Data page
base_url = "https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page"

# Send a GET request to the page
response = requests.get(base_url)
if response.status_code != 200:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
    exit()

# Parse the page content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all links ending with .parquet or .csv
download_links = []
for link in soup.find_all('a', href=True):
    href = link['href']
    if href.endswith('.parquet') or href.endswith('.csv'):
        # Handle relative URLs
        if href.startswith('/'):
            full_url = f"https://www.nyc.gov{href}"
        else:
            full_url = href
        download_links.append(full_url)

# Create a directory to store downloaded files
os.makedirs('tlc_trip_data', exist_ok=True)

# Download each file
for url in download_links:
    file_name = os.path.join('tlc_trip_data', url.split('/')[-1])
    print(f"Downloading {file_name}...")
    file_response = requests.get(url)
    if file_response.status_code == 200:
        with open(file_name, 'wb') as f:
            f.write(file_response.content)
        print(f"Downloaded {file_name}")
    else:
        print(f"Failed to download {url}. Status code: {file_response.status_code}")
```
